# add above code in main model training file and then run it

```python
#change col name as classcode & keep only classcode value
df_final['Comments'] = df_final['Comments'].str.replace('ClassCode:', '')
df_final = df_final.rename(columns={'Comments': 'ClassCode'})
df_final
```

```python
#remove - from zipcode value
df_final["ZipCode"] = df_final["ZipCode"].str.replace("-", "")
df_final
```

```python
df_final = df_final[pd.to_numeric(df_final['LossAmount'], errors='coerce').notna()]
df_final
```

```python
def extract_amount(value):
    match = re.search(r'(\d+\.\d+)', str(value))
    if match:
        return float(match.group(1))
    else:
        return None
```

```python
df_final['LossAmount'] = df_final['LossAmount'].apply(extract_amount)
df_final = df_final.dropna(subset=['LossAmount'])
df_final
```

```python
# in limit and deductible column, separate the alphabetical and numeric value in two separate columns
df_final['LossAmount'].fillna(0, inplace=True)
# df['Deductible'] = df['Deductible'].str.extract(r'(\d+\.\d+)').astype(float)
# df['Limit'] = df['Limit'].str.extract(r'(\d+\.\d+)').astype(float)
df_final['limit_type'] = df_final['Limit'].str.extract(r'([a-zA-Z]+)')
df_final['Limit_val'] = df_final['Limit'].str.extract(r'(\d+)')
df_final['Deductible_type'] = df_final['Deductible'].str.extract(r'([a-zA-Z]+)')
df_final['Deductible_val'] = df_final['Deductible'].str.extract(r'(\d+)')
df_final
```

```python
df_final['limit_type'].unique()
```

```python
df_final=df_final.drop(columns=["Limit", "Deductible", "YearBuilt"])
df_final
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
col_list = ["Jurisdiction", "CoverageType", "InjuryCause", "ClassCode", "ConstructionTypeDesc","ZipCode","limit_type","Deductible_type"]
for col in col_list:
    df_final[col]= label_encoder.fit_transform(df_final[col])
df_final
```

```python
columns_to_string = ['Jurisdiction', 'CoverageType', 'InjuryCause', 'ClassCode', 'ConstructionTypeDesc', 'ZipCode', 'limit_type',
                    'Deductible_type']
df_final[columns_to_string] = df_final[columns_to_string].astype(str)
columns_to_int = ['Stories', 'Deductible_val']
df_final.fillna(0, inplace=True)
df_final[columns_to_int] = df_final[columns_to_int].astype('int')
df_final['Limit_val'] = df_final['Limit_val'].astype('double')
#df['YearBuilt'] = df['YearBuilt'].astype('double')
df_final
```

```python
from sklearn.model_selection import KFold
from lightgbm import Dataset
```

```python
num_folds = 8
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)
y_pred = np.zeros(len(X_train))
```

```python
params = {
    'objective': 'regression',
    'metric': ['l1', 'l2'],
    'learning_rate': 0.005,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.7,
    'bagging_freq': 10,
    'verbose': 0,
    'max_depth': 8,
    'num_leaves': 128,
    'max_bin': 512,
    'num_boost_round': 100000,
    'early_stopping_round': 500,
    'verbosity': 100,
}
```

```python
for train_idx, val_idx in kf.split(X_train):
    X_train_fold, X_val_fold = X_train.iloc[train_idx], X_train.iloc[val_idx]
    y_train_fold, y_val_fold = y_train.iloc[train_idx], y_train.iloc[val_idx]
    train_data = lgb.Dataset(X_train_fold, label=y_train_fold)
    valid_data = lgb.Dataset(X_val_fold, label=y_val_fold, reference=train_data)
```
